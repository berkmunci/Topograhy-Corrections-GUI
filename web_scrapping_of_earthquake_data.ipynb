```python
#RUN FOR UPDATE DATABASE

import pandas as pd 
from tqdm import tqdm
import numpy as np
import urllib.request
from pprint import pprint
from html_table_parser.parser import HTMLTableParser
import pandas as pd
 
#1) DATA COLLECTION PHASE
def url_get_contents(url):
 
    req = urllib.request.Request(url=url)
    f = urllib.request.urlopen(req)
 
    return f.read()
 
xhtml = url_get_contents('http://www.koeri.boun.edu.tr/scripts/lst8.asp').decode("ISO-8859-1")

from bs4 import BeautifulSoup

#2) DATA PREPERATION PHASE
soup = BeautifulSoup(xhtml)

pre = soup.find_all('pre')[-1]
a=pre.text.strip()
dd1=a[584:].split("\r\n")

for i in tqdm(range(len(dd1))):
    dd2=dd1[i].split()
    date.append(dd2[0])
    time.append(dd2[1])
    sorter.append(dd2[0].replace(".","")+dd2[1].replace(":",""))
    valid.append(dd2[0].replace(".","")+dd2[1].replace(":","")+dd2[2].replace(".","")+dd2[3].replace(".",""))
    lat.append(dd2[2])
    lon.append(dd2[3])
    depth.append(dd2[4])
    Md.append(dd2[5])
    Ml.append(dd2[6])
    Mw.append(dd2[7])
    d_area=str(''.join(dd2[8:])).replace("Ýlksel","")
    area.append(d_area.replace("("+d_area[d_area.find("(")+1:d_area.find(")")]+")",""))
    city.append(d_area[d_area.find("(")+1:d_area.find(")")])


eq=pd.DataFrame(data=np.transpose([valid,date,time,lat,lon,depth,Md,Ml,Mw,area,city,sorter]),columns=["valid","date","time","lat","lon","depth_km","Md","Ml","Mw","area","city","sorter"])
eq=eq.drop_duplicates(['valid'])
eq=eq.sort_values(by='sorter',ascending=False).reset_index(drop=True)
eq=eq.drop(["sorter"],axis=1)

if len(eq["valid"].unique().tolist())==len(eq):
    print("WORKED! (There is no duplicated data)")
    display(eq)
else:
    print("Duplicated data is found.")
```
